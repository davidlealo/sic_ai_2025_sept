```python
import seaborn as sns

# Load the mpg dataset from seaborn
mpg_df = sns.load_dataset('mpg')

# Display the first few rows of the dataframe
display(mpg_df.head())
```

```python
mpg_df['origin'].unique()
```

```python
mpg_df.describe()
```

```python
# Select numerical columns, excluding 'model_year' which is more of a categorical feature here
numerical_cols = mpg_df.select_dtypes(include=['float64', 'int64']).columns.tolist()
# Removing 'model_year' as it represents a year and not a continuous numerical feature for scaling purposes in this context
numerical_cols.remove('model_year')

print("Numerical columns selected for scaling:")
print(numerical_cols)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
minmax_scaler = MinMaxScaler()

# Apply Min-Max scaling to the selected numerical columns
mpg_df_minmax_scaled = mpg_df.copy()
mpg_df_minmax_scaled[numerical_cols] = minmax_scaler.fit_transform(mpg_df[numerical_cols])

print("DataFrame after Min-Max Scaling:")
display(mpg_df_minmax_scaled.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
standard_scaler = StandardScaler()

# Apply Standard scaling to the selected numerical columns
mpg_df_standard_scaled = mpg_df.copy()
mpg_df_standard_scaled[numerical_cols] = standard_scaler.fit_transform(mpg_df[numerical_cols])

print("DataFrame after Standard Scaling:")
display(mpg_df_standard_scaled.head())
```

### Matching the Same Measurement Unit Equally

```python
df = pd.read_csv('./auto-mpg.csv', header=None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model year', 'orign', 'name']
print(df.head(3))
print('\n')
```

```python
mpg_to_kpl = 1.60934 / 3.78541
mpg_to_kpl
```

```python
df['kpl'] = df['mpg'] * mpg_to_kpl
```

```python
df
```

```python
df['kpl'] = df['kpl'].round(2)
df
```

```python
print(df.dtypes)
print('\n')
```

```python
print(df['orign'].unique())
print('\n')
```

```python
print(df['horsepower'].unique())
print('\n')
```

```python
# Count the number of '?' values in the 'horsepower' column
num_missing_horsepower = df['horsepower'].eq('?').sum()

print(f"Number of '?' values in the 'horsepower' column: {num_missing_horsepower}")
```

```python
df.info()
```

```python
# Display the rows where the 'horsepower' column is equal to '?'
display(df[df['horsepower'] == '?'])
```

```python
df.horsepower.describe()
```

```python
df.describe()
```

```python
import numpy as np

# Replace '?' values in the 'horsepower' column with 0
df['horsepower'] = df['horsepower'].replace('?', 0)

# Verify the replacement
print("Number of '?' values after replacement:")
print(df['horsepower'].eq('?').sum())
```

```python
df['horsepower'] = df['horsepower'].astype('float')
```

```python
df.info()
```

```python
df.horsepower.describe()
```

```python
# Get the frequency of the value 102 in the 'horsepower' column
frequency_102 = df[df['horsepower'] == 102].shape[0]
print(f"Frequency of horsepower = 102: {frequency_102}")

# Get the frequency of the value 92 in the 'horsepower' column
frequency_92 = df[df['horsepower'] == 92].shape[0]
print(f"Frequency of horsepower = 92: {frequency_92}")
```

```python
df['orign'].replace({1:'USA', 2:'EU', 3:'JAPAN'}, inplace=True)

print(df['orign'].unique())
print(df['orign'].dtypes)
print('\n')
```

```python
df['orign'] = df['orign'].astype('category')
print(df['orign'].dtypes)

df['orign'] = df['orign'].astype('str')
print(df['orign'].dtypes)
```
