### Manipulaci칩n del DataFrame

## **Ordenaci칩n:**  
Es posible ordenar un DataFrame utilizando una o m치s columnas. Para ello, se utiliza el m칠todo **`sort_values()`**, que permite ordenar un DataFrame o una Serie en funci칩n de los valores de una o m치s columnas (o del 칤ndice, en el caso de Series).

#### 游늷 Sintaxis simple
```python
DataFrame.sort_values(by, axis=0, ascending=True, inplace=False)
```
*   `by`: *str or list of str*. Especifica la columna o lista de columnas por las que se desea ordenar.
*   `axis`(0 = filas, 1=columas): Define el eje sobre el cual se va a ordenar. Por defecto es 0.
*   `ascending`: *bool or list of bools*, default True. Indica si el orden debe ser ascendente (True) o descendente (False).
*   `inplace`: *bool*, default False. Si es True, modifica el DataFrame original; si es False, devuelve una copia ordenada.
#### 游늷 Sintaxis completa
```python
DataFrame.sort_values(
    by,    axis=0,    ascending=True,    inplace=False,    
    kind='quicksort',    na_position='last',    ignore_index=False,    key=None
)
```
Revisar: [pandas.DataFrame.sort_values](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html) para m치s informaci칩n.


>Los ejemplos a continuaci칩n son con `axis = 0`. Es decir ordena las filas seg칰n los valores de una o m치s columnas


````python
import pandas as pd
import numpy as np

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan', 'Fiona'],
    'age': [28, 34, np.nan, 45, 30, 25],
    'gender': ['F', 'M', 'M', 'F', 'M', 'F'],
    'bloodtype': ['A', 'B', 'O', 'AB', 'A', 'A'],
    'weight': [55.0, np.nan, 68.0, 62.0, 85.0, 51.5],
    'height': [162, 175, 168, np.nan, 180, 160]
}

pacientes_df = pd.DataFrame(data)
````

````python
#Ordenar por un valor por defecto ser치 ascendente
pacientes_df.sort_values(by='age')
````

````python
# Ordenar en en forma descendente
pacientes_df.sort_values(by='bloodtype', ascending=False)
````

````python
# Ordenar utilizando dos columnas
pacientes_df.sort_values(by=['bloodtype','gender'], ascending=[False,False])
````

````python
pacientes_df.sort_values(by='height')
````

````python
pacientes_df.sort_values(by='height', na_position='first')
````

> Utilizando `axis = 1`. Es decir ordena las columnas seg칰n los valores de una o m치s filas
##### 丘멆잺 Advertencia: Error al ordenar con `axis=1` si hay mezcla de tipos

Cuando se utiliza `sort_values()` con `axis=1`, se intenta **ordenar las columnas** utilizando los valores de una **fila espec칤fica**. Esto puede generar un error si esa fila contiene **tipos de datos incompatibles**, como n칰meros (`float`, `int`) y cadenas de texto (`str`), ya que **no pueden compararse entre s칤**.

#### 游빍 Ejemplo que genera error

```python
# DataFrame con mezcla de tipos (num칠ricos y strings)
pacientes_df.sort_values(by=0, axis=1)
```
游눤 Error esperado:
```txt
TypeError: ufunc 'greater' did not contain a loop with signature matching types
(<class 'numpy.dtype[float64]'>, <class 'numpy.dtype[str_]'>) -> None
```

````python
# Opci칩n para evitar el error
# Solo columnas num칠ricas para evitar errores
numericas = pacientes_df.select_dtypes(include='number')

# Ordenar columnas seg칰n los valores de la fila 0
ordenadas1 = numericas.sort_values(by=3, axis=1, na_position='first')
ordenadas2 = numericas.sort_values(by=3, axis=1)
print(ordenadas1)
print('-'*26)
print(ordenadas2)
````

## **Indexaci칩n Jerarquica con MultiIndex:**

Un MultiIndex en Pandas es una estructura que permite tener 칤ndices jer치rquicos en las filas o columnas de un DataFrame o Series. En lugar de un 칰nico 칤ndice, puedes tener m칰ltiples niveles de etiquetas, lo que facilita la organizaci칩n y el acceso a datos complejos.

**Para qu칠 sirve un MultiIndex**

* Organizar mejor la informaci칩n

* Puedes agrupar filas por categor칤as y subcategor칤as.

````python
# Importamos las librer칤as necesarias: pandas para manejo de datos y numpy para generar n칰meros aleatorios
import pandas as pd
import numpy as np
````

````python
# Definimos los nombres de las columnas del DataFrame
my_header = ['A', 'B', 'C']

# Creamos la primera parte del 칤ndice (nivel externo) con etiquetas de grupo: 'G1', 'G2', 'G3', cada una repetida dos veces
my_index_out = ['G1']*2 + ['G2']*2 + ['G3']*2
# Creamos la segunda parte del 칤ndice (nivel interno) con sub-etiquetas: '1', '2' repetidas tres veces
my_index_in = ['1', '2']*3

# Combinamos los dos niveles de 칤ndice en una lista de tuplas, una por cada fila
my_index_zipped = list(zip(my_index_out, my_index_in))

# Creamos un MultiIndex a partir de la lista de tuplas
my_multi_index = pd.MultiIndex.from_tuples(my_index_zipped)

# Creamos el DataFrame con:
# - 6 filas (una por cada tupla del 칤ndice)
# - 3 columnas ('A', 'B', 'C')

# Aplicaci칩n de MultiIndex en el 칤ndice
df = pd.DataFrame(data=np.random.randn(6, 3), # valores aleatorios (filas, columnas)
                  index=my_multi_index, ## 游댮 AQU칈 est치 la diferencia
                  columns=my_header)
df
````

````python
# Accede a todas las filas que tienen 'G1' en el primer nivel del 칤ndice (nivel externo del MultiIndex)
# Devuelve un sub-DataFrame con las subetiquetas del segundo nivel (por ejemplo: '1' y '2')
df.loc['G1']
````

````python
# Indexing followed by another indexing.
# Esto selecciona todas las filas del DataFrame que tienen 'G1' como valor en el primer nivel
df.loc['G1'].loc['1']
# Sobre el sub-DataFrame anterior, accedemos a la fila con 칤ndice '1'.
#G1	1	1.179609	-1.334867	0.476122
````

````python
# Indexing followed by another indexing.
df.loc['G1'].loc['1','B']
# Sigue la l칩gica del ejemplo anterior pero accede a la celda espec칤fica que est치 en:
# La fila con 칤ndice '1' y la columna con nombre 'B'
````

El m칠todo `groupby()` de pandas se utiliza para **agrupar filas** de un DataFrame seg칰n una o m치s columnas que tengan **valores iguales**.

Esto permite **resumir** la informaci칩n de forma clara y segmentada por grupos.

## **Pivotante**
Pivotar un DataFrame significa reorganizar los datos:
Pasas de un formato largo (filas con muchas repeticiones) a un formato ancho (filas resumidas con columnas m치s informativas).
### 쯇ara qu칠 sirve pivotar?

- **Comparar datos f치cilmente** al convertir valores de una columna en columnas separadas.
- **Preparar datos para gr치ficos** que requieren formato ancho.
- **Resumir y organizar datos agrupados** para una lectura r치pida.
- **Facilitar la creaci칩n de informes** claros y estructurados.

````python
import pandas as pd

my_dict = {
    "Size": ["L", "L", "M", "M", "M", "S", "S", "S", "S"],
    "Type": ["A", "A", "A", "B", "B", "A", "A", "B", "B"],
    "Location": ["L1", "L1", "L1", "L2", "L2", "L1", "L2", "L2", "L1"],
    "A": [1, 2, 2, 3, 3, 4, 5, 6, 7],
    "B": [2, 4, 5, 5, 6, 6, 8, 9, 9]
}

df = pd.DataFrame(my_dict)
df
````

````python
# 칈ndice por 'Size' y 'Type'. Columnas por 'Location'. Valores proporcionados por la columna 'B'.
dfr = pd.pivot_table(df, index=['Size', 'Type'], columns='Location', values='B')
dfr
````

## Advertencia al usar `aggfunc=np.median` en `pivot_table`

Cuando usamos el c칩digo de la presentaci칩n:

```python

pd.pivot_table(df, index=['Size', 'Type'], columns='Location', values='B', aggfunc=np.median, fill_value=0)
```
Arroja una advertencia
```txt
FutureWarning: The provided callable <function median> is currently using DataFrameGroupBy.median.. In a future version of pandas, the provided callable will be used directly. To keep current behavior pass the string "median" instead.
```
Lo que quiere decir que en el futuro, Pandas usar치 directamente la funci칩n que le pases para calcular la mediana, lo que podr칤a cambiar ligeramente los resultados si la funci칩n y el m칠todo interno funcionan diferente.

````python
# Codigo ex_0109
# Visualizaci칩n b치sica con Matplotlib
# Line plot / Grafico de lineas
x = np.linspace(0, 10, 100)                         # Valores equiespaciados entre 0 y 10
y = np.sin(x)                                       # Funci칩n seno
plt.plot(x, y, color='orange', linewidth=2)         # L칤nea continua
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='blue', linestyle=':', linewidth=2)   # L칤nea punteada
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='red', linestyle='-.', linewidth=2)   # L칤nea punto-guion
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='green', linestyle='--', linewidth=2) # L칤nea discontinua
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()

# Ejemplo de error: linestyle no acepta 'steps'
# plt.plot(x, y, color='purple', linestyle='steps')

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='purple', linestyle='-', linewidth=2) # L칤nea continua
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.xlim([0, 5])                                    # L칤mite del eje X
plt.ylim([-2, +2])                                  # L칤mite del eje Y
plt.show()
````

````python
# Codigo ex_0109
# Visualizaci칩n b치sica con Matplotlib
# Scatterplot / Grafica de dispersion
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='red', marker='o', linestyle='none', markersize=5, alpha=0.5)
plt.title('SCATTER PLOT')
plt.show()

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='#FF8C00', marker='o', linestyle='none', markersize=5, alpha=0.5)  # Color en RGB
plt.title('SCATTER PLOT')
plt.show()

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, marker='o', linestyle='none', markersize=15,
         markerfacecolor='yellow', markeredgecolor='purple', markeredgewidth=2)
plt.title('SCATTER PLOT')
plt.show()
````
