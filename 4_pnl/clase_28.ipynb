# Explicación Detallada de los Hiperparámetros de `TfidfVectorizer`

La clase `TfidfVectorizer` de scikit-learn convierte una colección de documentos sin procesar en una matriz de características TF-IDF (Frecuencia de Término - Frecuencia Inversa de Documento).

## 1. Parámetros de Preprocesamiento y Análisis

Estos controlan la lectura, limpieza y segmentación del texto.

* **`input`**
    * **Valores:** `'content'` (predeterminado), `'filename'`, `'file'`.
    * **Descripción:** Define el tipo de entrada: cadenas de texto (`'content'`), rutas de archivo (`'filename'`), u objetos con método `.read()` (`'file'`).
* **`encoding`**
    * **Valor Predeterminado:** `'utf-8'`.
    * **Descripción:** Codificación utilizada para decodificar bytes o archivos.
* **`decode_error`**
    * **Valores:** `'strict'` (predeterminado), `'ignore'`, `'replace'`.
    * **Descripción:** Manejo de errores de decodificación de bytes.
* **`strip_accents`**
    * **Valores:** `'ascii'`, `'unicode'`, `None` (predeterminado).
    * **Descripción:** Elimina acentos y normaliza caracteres. `None` deshabilita la función.
* **`lowercase`**
    * **Valor Predeterminado:** `True`.
    * **Descripción:** Convierte todo el texto a minúsculas antes de la tokenización.
* **`preprocessor`**
    * **Valor Predeterminado:** `None`.
    * **Descripción:** Función personalizada para sobrescribir la etapa de limpieza de la cadena.
* **`tokenizer`**
    * **Valor Predeterminado:** `None`.
    * **Descripción:** Función personalizada para sobrescribir el paso de división en *tokens*. Solo se usa si `analyzer == 'word'`.
* **`analyzer`**
    * **Valores:** `'word'` (predeterminado), `'char'`, `'char_wb'`, o un *callable*.
    * **Descripción:** Define si las características son n-gramas de palabras, de caracteres o de caracteres dentro de los límites de palabras.
* **`stop_words`**
    * **Valores:** `'english'`, `list`, `None` (predeterminado).
    * **Descripción:** Lista de palabras vacías a excluir. Si es `'english'`, usa una lista predefinida.
* **`token_pattern`**
    * **Valor Predeterminado:** `r”(?u)\b\w\w+\b”`.
    * **Descripción:** Expresión regular para definir lo que se considera un *token*. Solo se aplica si `analyzer == 'word'`.
* **`ngram_range`**
    * **Valor Predeterminado:** `(1, 1)`.
    * **Descripción:** Tupla `(min_n, max_n)` que define el rango de tamaño de los n-gramas a extraer. Ej: `(1, 2)` para unigramas y bigramas.

## 2. Parámetros de Construcción del Vocabulario

Estos parámetros controlan la selección de términos que formarán las columnas de la matriz.

* **`max_df`**
    * **Valor Predeterminado:** `1.0`.
    * **Descripción:** Umbral para ignorar términos con una Frecuencia de Documento (DF) **superior** al límite. Puede ser una proporción (`float` entre 0.0 y 1.0) o un conteo absoluto (`int`). Útil para eliminar palabras vacías específicas del corpus.
* **`min_df`**
    * **Valor Predeterminado:** `1`.
    * **Descripción:** Umbral para ignorar términos con una Frecuencia de Documento (DF) **inferior** al límite. Puede ser una proporción o un conteo absoluto. Ayuda a eliminar términos muy raros.
* **`max_features`**
    * **Valor Predeterminado:** `None`.
    * **Descripción:** Si se especifica, limita el vocabulario a las `max_features` más frecuentes.
* **`vocabulary`**
    * **Valor Predeterminado:** `None`.
    * **Descripción:** Permite proporcionar un vocabulario fijo predefinido, anulando los filtros `max_df`, `min_df` y `max_features`.
* **`binary`**
    * **Valor Predeterminado:** `False`.
    * **Descripción:** Si es `True`, la Frecuencia de Término (TF) se establece en 1 para cualquier término que aparezca, ignorando el conteo real.

## 3. Parámetros de Ponderación y Normalización TF-IDF

Estos controlan cómo se calculan y escalan los pesos finales de la matriz.

* **`dtype`**
    * **Valor Predeterminado:** `numpy.float64`.
    * **Descripción:** Tipo de datos de la matriz de características de salida.
* **`norm`**
    * **Valores:** `'l2'` (predeterminado), `'l1'`, `None`.
    * **Descripción:** Normalización aplicada a cada vector de documento. `'l2'` es la norma euclidiana, comúnmente usada para la similitud del coseno. `None` desactiva la normalización.
* **`use_idf`**
    * **Valor Predeterminado:** `True`.
    * **Descripción:** Habilita el re-pesaje por Frecuencia Inversa de Documento (IDF). Si es `False`, se calcula solo la Frecuencia de Término (TF) normalizada.
* **`smooth_idf`**
    * **Valor Predeterminado:** `True`.
    * **Descripción:** Suaviza los pesos IDF añadiendo +1 a las frecuencias de documento. Esto evita la división por cero para términos raros y suaviza los pesos.
* **`sublinear_tf`**
    * **Valor Predeterminado:** `False`.
    * **Descripción:** Aplica un escalado sublineal a la TF, reemplazando el conteo de términos con **$1 + \log(\text{TF})$** para mitigar el impacto de un conteo muy alto en un solo documento.

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

my_docs = ["The economic slowdown is becoming more severe",
           "The movie was simply awesome",
           "I like cooking my own food",
           "Samsung is announcing a new technology",
           "Machine Learning is an example of awesome technology",
           "All of us were excited at the movie",
           "We have to do more to reverse the economic slowdown"]

# Create a CountVectorizer object
vectorizer = CountVectorizer()

# Create the DTM
dtm = vectorizer.fit_transform(my_docs)

# Get the feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Sum the frequencies of each word across all documents
word_frequencies = dtm.sum(axis=0)

# Create a dictionary of word frequencies
frequencies_dict = dict(zip(feature_names, word_frequencies.tolist()[0]))

# Create a pandas Series for better display
frequencies_series = pd.Series(frequencies_dict)

# Sort the frequencies in descending order
sorted_frequencies = frequencies_series.sort_values(ascending=False)

# Print the word frequencies
print("Word Frequencies:")
print(sorted_frequencies)
```

```python
import numpy as np
import warnings
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition  import TruncatedSVD
warnings.filterwarnings('ignore')
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
```

```python
# The data.
my_docs = ["The economic slowdown is becoming more severe",
           "The movie was simply awesome",
           "I like cooking my own food",
           "Samsung is announcing a new technology",
           "Machine Learning is an example of awesome technology",
           "All of us were excited at the movie",
           "We have to do more to reverse the economic slowdown"]
```

```python
my_docs = [x.lower() for x in my_docs]
```

```python
my_stop_words = ['us', 'like']
```

```python
vectorizer = TfidfVectorizer(max_features = 15, min_df = 1, max_df = 3, stop_words = stopwords.words('english') + my_stop_words)
X = vectorizer.fit_transform(my_docs).toarray()
```

```python
X[0]
```

```python
import pandas as pd

# Create a DataFrame from the matrix X and feature names
X_df = pd.DataFrame(X, columns=vectorizer.get_feature_names_out())

# Display the DataFrame
display(X_df)
```

```python
# View the features.
features = vectorizer.get_feature_names_out()
print(features)
```

```python
# Apply the truncated SVD:

n_topics = 4
svd = TruncatedSVD(n_components=n_topics, n_iter=100)
svd.fit(X)
```

```python
# get the V^t matrix.
vt = svd.components_
vtabs = np.abs(vt)
```

```python
# Display the vt matrix
print("V^t matrix:")
print(vt)
```

```python
vt.shape
```

```python
n_top = 3
for i in range(n_topics):
    topic_features = [features[idx] for idx in np.argsort(-vtabs[i,:])]   # argsort() shows the sorted index.
    topic_features_top = topic_features[0:n_top]
    if i == 0:
        topic_matrix = [topic_features_top]                    # list의 list 만들 준비!
    else:
        topic_matrix.append(topic_features_top)
```

```python
# Show the top features for each topic.
topic_matrix
```

```python
topic_names = ['Economy', 'Movie','Technology', 'Cuisine'] # In view of the top features, we can name the topics.
```

```python
n_docs = len(my_docs)
for i in range(n_docs):
    score_pick = 0
    topic_pick = 0
    tokennized_doc = nltk.word_tokenize(my_docs[i])
    for j in range(n_topics):
        found = [ x in topic_matrix[j] for x in tokennized_doc ]
        score = np.sum(found)
        if (score > score_pick):
            score_pick = score
            topic_pick = j
    print("Document " + str(i+1) + " = " + topic_names[topic_pick])
```

```python
import pandas as pd

# Assuming topic_names and my_docs are defined from previous steps
# If not, make sure they are available in the environment

# Create a list of document and topic name pairs
results = []
for i in range(len(my_docs)):
    # Find the predicted topic index for the document
    topic_pick = -1
    max_score = -1
    tokenized_doc = nltk.word_tokenize(my_docs[i].lower()) # Tokenize and lowercase the document

    for j in range(len(topic_matrix)):
        # Calculate a score based on the presence of top words for each topic
        score = sum(word in topic_matrix[j] for word in tokenized_doc)
        if score > max_score:
            max_score = score
            topic_pick = j

    # Get the topic name
    assigned_topic = topic_names[topic_pick]

    results.append({'Document': my_docs[i], 'Assigned Topic': assigned_topic})

# Create a pandas DataFrame
results_df = pd.DataFrame(results)

# Display the DataFrame
display(results_df)
```

## Explicación Detallada de los Resultados de LSA y SVD

Hemos aplicado el Análisis Semántico Latente (LSA) utilizando la Descomposición de Valores Singulares Truncada (Truncated SVD) a nuestros documentos (`my_docs`). Aquí explicamos los resultados:

1.  **Construcción de la Matriz TF-IDF (`X`)**:
    *   Primero, convertimos nuestros documentos en una Matriz de Frecuencia de Término - Frecuencia Inversa de Documento (TF-IDF). Esta matriz (`X_df` en el notebook) representa la importancia de cada palabra en cada documento. Las filas corresponden a los documentos y las columnas a las palabras (características). Los valores en la matriz son los pesos TF-IDF, donde un valor alto indica que una palabra es importante y relevante para un documento específico.

2.  **Aplicación de Truncated SVD**:
    *   La matriz TF-IDF (`X`) es a menudo dispersa y de alta dimensión. SVD (o Truncated SVD en este caso, que es más eficiente para matrices dispersas) es una técnica de reducción de dimensionalidad. Descompone la matriz original `X` en tres matrices: $U$, $\Sigma$ (Sigma) y $V^T$.
    *   La matriz $V^T$ (mostrada como `vt` en el notebook) es particularmente interesante para entender los temas. Tiene dimensiones (número de temas) x (número de palabras). Cada fila de $V^T$ corresponde a un tema latente, y los valores en esa fila indican la fuerza de asociación entre cada palabra y ese tema. Valores más altos (en magnitud, ya que pueden ser negativos) significan una mayor relevancia de la palabra para ese tema.

3.  **Identificación de Temas**:
    *   Al examinar las palabras con los pesos más altos en cada fila de la matriz $V^T$ (utilizando `vtabs` que son los valores absolutos para considerar la magnitud independientemente del signo), podemos inferir el tema que representa cada fila.
    *   En nuestro caso, hemos identificado 4 temas (`n_topics = 4`). Al observar las 3 palabras principales (`n_top = 3`) con los mayores pesos para cada tema (mostrado en `topic_matrix`), pudimos asignar nombres descriptivos a cada tema: 'Economy', 'Movie', 'Technology', y 'Cuisine'.

4.  **Clasificación de Documentos por Tema**:
    *   Finalmente, clasificamos cada documento (`my_docs`) en uno de los temas identificados. Esto se hizo observando qué tema tenía la mayor superposición de palabras con las palabras presentes en cada documento.
    *   La tabla resultante (`results_df`) muestra cada documento original y el tema al que fue asignado. Por ejemplo:
        *   Documentos sobre "slowdown" y "economic" fueron asignados al tema 'Economy'.
        *   Documentos con palabras como "movie", "awesome" y "excited" fueron asignados al tema 'Movie'.
        *   Documentos con "technology", "new" y "samsung" fueron al tema 'Technology'.
        *   Documentos sobre "cooking" y "food" fueron al tema 'Cuisine'.

En resumen, LSA con SVD nos ha permitido descubrir los temas subyacentes en nuestra colección de documentos y clasificar cada documento según su tema principal, basándose en la co-ocurrencia y ponderación de las palabras.

## Interacción del Algoritmo LDA con Bayes y la Distribución Multinomial

El Algoritmo de Asignación Latente de Dirichlet (LDA) es un modelo generativo probabilístico que se utiliza para descubrir temas en una colección de documentos. Su formulación se basa fundamentalmente en principios Bayesianos y utiliza la distribución multinomial para modelar la aparición de palabras en los documentos.

Aquí se explica la interacción:

1.  **Modelo Generativo de LDA:** LDA postula un proceso generativo para cada documento en el corpus. Este proceso describe cómo se "crea" un documento si se conocen los temas subyacentes y las distribuciones de palabras dentro de esos temas.

2.  **Distribuciones de Dirichlet (Priors Bayesianos):**
    *   LDA utiliza distribuciones de Dirichlet como **distribuciones a priori** (priors) sobre dos aspectos clave:
        *   **Distribución de temas por documento ($\theta$):** Para cada documento, LDA asume que la proporción de cada tema en ese documento sigue una distribución de Dirichlet. Esto significa que, a priori, esperamos que los temas tengan una cierta distribución de probabilidad en cada documento. El parámetro de esta Dirichlet (generalmente $\alpha$) influye en si esperamos que los documentos tengan pocos temas predominantes (valores bajos de $\alpha$) o una mezcla más uniforme de temas (valores altos de $\alpha$).
        *   **Distribución de palabras por tema ($\phi$):** Para cada tema, LDA asume que la proporción de cada palabra en ese tema sigue una distribución de Dirichlet. Esto significa que, a priori, esperamos que ciertas palabras tengan una cierta distribución de probabilidad dentro de cada tema. El parámetro de esta Dirichlet (generalmente $\beta$) influye en si esperamos que los temas se definan por pocas palabras muy probables (valores bajos de $\beta$) o por una mezcla más uniforme de palabras (valores altos de $\beta$).
    *   La elección de la distribución de Dirichlet como prior es conveniente porque es la **conjugada** de la distribución multinomial. Esto simplifica los cálculos en la inferencia.

3.  **Distribución Multinomial (Likelihood):**
    *   Dado un documento y la distribución de temas para ese documento ($\theta$), y dada una palabra y la distribución de palabras para cada tema ($\phi$), la probabilidad de observar esa palabra en ese documento se modela utilizando la **distribución multinomial**.
    *   Específicamente, para cada palabra en un documento:
        *   Primero, se elige un tema de la distribución de temas del documento ($\theta$). Esto es una muestra de una distribución multinomial sobre los temas.
        *   Luego, dado el tema elegido, se elige una palabra de la distribución de palabras de ese tema ($\phi$). Esto es una muestra de una distribución multinomial sobre el vocabulario.
    *   La probabilidad de todo el corpus observado es el producto de las probabilidades multinomiales de cada palabra en cada documento, condicionada a las distribuciones de Dirichlet para temas y palabras.

4.  **Inferencia Bayesiana:**
    *   El objetivo de LDA es la **inferencia**: dado el corpus de documentos observados, queremos encontrar las distribuciones de temas por documento ($\theta$) y las distribuciones de palabras por tema ($\phi$) que mejor explican los datos.
    *   LDA no calcula un único valor para $\theta$ y $\phi$ (como en un enfoque de máxima verosimilitud), sino que busca la **distribución posterior** de estos parámetros, dada la evidencia de los documentos. Esto es el corazón del enfoque Bayesiano: combinar las creencias a priori (las distribuciones de Dirichlet) con la evidencia de los datos (la probabilidad multinomial de las palabras observadas) para obtener la distribución posterior.
    *   La inferencia en LDA a menudo se realiza utilizando técnicas como el muestreo de Gibbs o la Inferencia Variacional Bayesiana, que son métodos para aproximar la distribución posterior que es difícil de calcular directamente.

**En resumen:**

LDA utiliza un marco Bayesiano al imponer distribuciones de Dirichlet como priors sobre las distribuciones de temas y palabras. La observación de las palabras en los documentos se modela mediante distribuciones multinomiales. El algoritmo infiere las distribuciones posteriores de temas y palabras combinando estos priors con la likelihood multinomial de los datos observados. La distribución de Dirichlet es clave porque es la prior conjugada de la multinomial, lo que facilita la inferencia en el modelo.

```python
import numpy as np
import warnings
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition  import LatentDirichletAllocation
warnings.filterwarnings('ignore')
```

```python
# The data.
my_docs = ["The economic slowdown is becoming more severe",
           "The movie was simply awesome",
           "I like cooking my own food",
           "Samsung is announcing a new technology",
           "Machine Learning is an example of awesome technology",
           "All of us were excited at the movie",
           "We have to do more to reverse the economic slowdown"]
```

```python
my_docs = [x.lower() for x in my_docs]
```

```python
vectorizer = CountVectorizer(max_features = 15, min_df = 1, max_df = 3, stop_words = list(ENGLISH_STOP_WORDS))
X = vectorizer.fit_transform(my_docs).toarray()
```

```python
# Show th DTM.
X
```

```python
# Size of X (=m x n). m = number of documents = 7 & n = number of features.
X.shape
```

```python
# View the features.
features = vectorizer.get_feature_names_out()
print(features)
```

```python
import pandas as pd

# Create a DataFrame from the matrix X and feature names
X_df = pd.DataFrame(X, columns=vectorizer.get_feature_names_out())

# Display the DataFrame
display(X_df)
```

```python
# Get the topics.
n_topics = 4
lda = LatentDirichletAllocation(n_components=n_topics, random_state=123)
my_docs_topic = lda.fit_transform(X)
```

```python
# row = document, column = topic.
my_docs_topic
```

```python
# Sum along the row has to give 1.
my_docs_topic.sum(axis=1)
```

```python
topic_composition = lda.components_
topic_composition.shape     # row = topic, column = feature (word).
```

```python
n_top = 3
for i in range(n_topics):
    topic_features = [features[idx] for idx in np.argsort(-topic_composition[i,:])]   # argsort() shows the sorted index.
    topic_features_top = topic_features[0:n_top]
    if i == 0:
        topic_matrix = [topic_features_top]                    # list의 list 만들 준비!
    else:
        topic_matrix.append(topic_features_top)
```

```python
# Show the top features for each topic.
topic_matrix
```

```python
# In view of the top features, we can name the topics.
topic_names = ['Technology', 'Cuisine', 'Movie','Economy']
```

```python
# The most probable topic is given directly by the LDA output.
n_docs = len(my_docs)
for i in range(n_docs):
    topic_pick = np.argmax(my_docs_topic[i,:])
    print("Document " + str(i+1) + " = " + topic_names[topic_pick])
```

```python
import pandas as pd

# Assuming my_docs, my_docs_topic, and topic_names are defined from previous steps

# Get the assigned topic index for each document
assigned_topic_indices = np.argmax(my_docs_topic, axis=1)

# Get the corresponding topic names
assigned_topic_names = [topic_names[idx] for idx in assigned_topic_indices]

# Create a pandas DataFrame
results_df_lda = pd.DataFrame({'Document': my_docs, 'Assigned Topic (LDA)': assigned_topic_names})

# Display the DataFrame
display(results_df_lda)
```
