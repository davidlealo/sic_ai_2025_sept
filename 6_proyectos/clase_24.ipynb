------------------------------------------------------------------------

## **3. Concepto de prototipado**

El **prototipado** es la segunda etapa del *espacio de la solución* y
consiste en **crear representaciones rápidas, económicas y
experimentales** que permitan: - Comunicar la idea a otros. - Iniciar
conversaciones con usuarios reales. - Cometer errores pequeños y
baratos. - Explorar alternativas antes de comprometer recursos
importantes.

El profesor destacó que **el error es deseable** en esta etapa: cada
fallo es una oportunidad para aprender y mejorar.

------------------------------------------------------------------------

## **4. Principios del prototipado**

1.  **Construir sin esperar claridad total:** empezar a hacer aunque el
    resultado no esté definido.\
2.  **Prototipar en escala pequeña:** comenzar con versiones simples y
    luego complejizar.\
3.  **Cada prototipo debe responder una pregunta:** por ejemplo:
    -   ¿Cuál es su apariencia?\
    -   ¿Cómo se sostiene o manipula?\
    -   ¿Qué funciones cumple?\
4.  **Mantener al usuario en el centro:** todo diseño debe considerar la
    interacción futura con quien lo probará.\
5.  **Registrar el proceso:** documentar con fotos, videos o capturas
    las pruebas, sobre todo la interacción usuario-prototipo.

------------------------------------------------------------------------

## **5. Tipos de prototipos**

-   **Físicos:** maquetas, modelos 3D, materiales reciclados, prototipos
    de hardware.\
-   **Digitales:** wireframes, mockups, aplicaciones en Figma o Flutter,
    simulaciones web.\
-   **De experiencia:** dramatizaciones, juegos de roles o storyboards
    que representen la interacción del usuario.

El docente insistió en que un **prototipo no tiene que ser funcional**:
basta con que permita **probar hipótesis** o comunicar la idea.

------------------------------------------------------------------------

## **6. Protocolo de creación de prototipos**

El proceso propuesto siguió cuatro pasos:

1.  **Definir qué será prototipado.**\
    Ejemplo: interfaz, flujo de usuario, algoritmo, o aspecto visual.\
2.  **Identificar la pregunta que busca responder.**\
    Ejemplo: "¿Entiende el usuario esta navegación?"\
3.  **Establecer el objetivo del prototipo.**\
    Ejemplo: probar la estética o la comprensión del dashboard.\
4.  **Elaborar un storyboard o guion de uso.**\
    Describir cómo se espera que el usuario interactúe con el prototipo
    (abrir la app, realizar una acción, etc.).

## **10. Concepto de evaluación**

El docente explicó que la **evaluación del prototipo** no es un examen,
sino una **oportunidad de aprendizaje**.

### **Propósitos:**

-   Recibir retroalimentación directa de los usuarios.\
-   Identificar errores y mejoras.\
-   Generar nuevos *insights* sobre el comportamiento de los usuarios.\
-   Redefinir el problema si es necesario.

**Regla clave:**\
\> "Construimos creyendo que tenemos la razón, y evaluamos pensando que
podemos estar equivocados."

------------------------------------------------------------------------

## **11. Estrategias para evaluar**

1.  **Observar la interacción sin intervenir.**\
    Permitir que el usuario explore el prototipo libremente.\
2.  **Registrar reacciones y comentarios.**\
    Notas, grabaciones o capturas visuales.\
3.  **Preguntar al final.**\
    Entrevista breve para conocer percepciones.\
4.  **Comparar alternativas.**\
    Mostrar dos versiones para descubrir preferencias.

------------------------------------------------------------------------

## **12. Herramienta: Matriz de Evaluación**

La **Matriz de Evaluación o Cuadrante Reflexivo** consta de cuatro
secciones:

  -----------------------------------------------------------------------
  Cuadrante                          Descripción
  ---------------------------------- ------------------------------------
  **Aspectos interesantes**          Hallazgos notables, incluso si algo
                                     no funcionó.

  **Críticas constructivas**         Opiniones del usuario sobre
                                     debilidades o mejoras.

  **Preguntas surgidas**             Dudas o reflexiones del equipo
                                     durante la prueba.

  **Ideas generadas**                Cambios o innovaciones que emergen
                                     del proceso.
  -----------------------------------------------------------------------

El análisis del cuadrante permite **redefinir el prototipo**, **idear
mejoras** y, si es necesario, **replantear el problema original**.

## **13. Caso práctico del docente**

El profesor compartió su experiencia con un proyecto real:\
una **plataforma educativa con IA integrada** para docentes.\
- **Prototipado inicial:** chat conversacional personalizado en GPT.\
- **Pruebas:** validación de estructura de consultas y flujo
conversacional.\
- **Iteraciones:** conexión con API de ChatGPT y despliegue en página
web.\
- **Aprendizaje:** el proceso requirió múltiples ciclos de prueba-error
antes de lograr una versión funcional (MVP).

Este ejemplo ilustró cómo **el ciclo iterativo de
idear--prototipar--evaluar--redefinir** se aplica también en entornos
profesionales.
