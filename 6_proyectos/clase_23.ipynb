# **Clase 23 -- Design Thinking: Espacio de la Solución**

## **1. Contexto general**

La **Clase 23** continúa el módulo de *Diseño de Proyectos*, centrado en
la metodología **Design Thinking**.\

2. **Prototipar**\
3. **Evaluar**

El objetivo de la clase fue **generar ideas creativas y viables** que

El profesor realizó un breve repaso de la **Clase 22**, recordando que

Hasta el momento, los grupos habían trabajado en las dos primeras
etapas: - **Empatizar**, mediante entrevistas, mapas de empatía y

------------------------------------------------------------------------

## **3. Introducción al espacio de la solución**

El **espacio de la solución** está compuesto por tres etapas
fundamentales: 1. **Idear** -- Generar una amplia gama de posibles
soluciones.\
2. **Prototipar** -- Crear representaciones o modelos de esas
soluciones.\
3. **Evaluar** -- Ponerlas a prueba con los usuarios y recoger
retroalimentación.

Estas fases son **iterativas**, lo que significa que pueden repetirse
múltiples veces hasta llegar a una versión más depurada y efectiva del
producto o servicio.

El profesor enfatizó que **idear no es buscar la mejor idea de
inmediato**, sino **abrir el pensamiento** para luego converger hacia
propuestas más refinadas.

------------------------------------------------------------------------

## **4. Etapa de ideación**

La sesión se centró especialmente en la **etapa de ideación**,
considerada la "primera patita" del proceso creativo.\
Se explicaron sus características clave:

-   Combina **pensamiento racional e intuitivo**.\
-   Prioriza **cantidad sobre calidad** en una primera fase.\
-   Requiere **colaboración, apertura y no juicio** entre los
    participantes.\
-   Se desarrolla mediante **preguntas guía** del tipo *¿Cómo
    podríamos...?*

### **Reglas para la ideación (Brainstorming)**

1.  Una sola conversación a la vez.\
2.  Respetar la voz de todos.\
3.  Construir sobre ideas ajenas.\
4.  Apoyar ideas "locas" o poco convencionales.\
5.  Mantenerse en el tema.\
6.  Priorizar cantidad sobre calidad.\
7.  Evitar juzgar o descartar ideas en esta etapa.

Estas normas fueron aplicadas en la plataforma **Miro**, donde los
equipos registraron sus ideas mediante notas visuales.

------------------------------------------------------------------------

## **5. Divergencia y convergencia**

El profesor explicó la diferencia entre ambos procesos: -
**Divergencia:** Fase expansiva, donde se generan muchas ideas sin
evaluar su factibilidad.\
- **Convergencia:** Fase reductiva, donde se seleccionan y refinan las
ideas más prometedoras.

Ambas son necesarias dentro del proceso creativo.\
Primero se explora libremente, y luego se filtra, combinando ideas o
agrupándolas en torno a criterios compartidos.

------------------------------------------------------------------------

## **6. Actividad grupal: generación de preguntas guía**

Cada grupo retomó el trabajo anterior y formuló **preguntas tipo "¿Cómo
podríamos...?"** relacionadas con su problemática.\
Estas preguntas debían ser **lo suficientemente amplias para permitir la
creatividad**, pero **suficientemente específicas para orientar la
solución**.

### **Ejemplos discutidos:**

-   *¿Cómo podríamos crear un sistema digital simple que permita
    coordinar tareas entre dueños y cuidadores de mascotas?*\
-   *¿Cómo podríamos fomentar la educación financiera de forma simple y
    motivante?*\
-   *¿Cómo podríamos optimizar las rutas de recolección de basura
    considerando la salud de los trabajadores?*

Los grupos presentaron sus borradores, debatieron sobre el grado de
especificidad y reformularon las preguntas en función de la
retroalimentación.

------------------------------------------------------------------------

## **7. Actividad práctica: brainstorming colaborativo**

Una vez definidas las preguntas guía, se realizó el ejercicio de
**brainstorming principal**.\
Cada grupo debía: 1. Generar al menos **20 ideas** posibles de
solución.\
2. Visualizarlas en **Miro** mediante notas o dibujos.\
3. Respetar las reglas del proceso creativo.\
4. Evitar críticas o filtros prematuros.

El profesor recordó que **las mejores ideas suelen emerger de
combinaciones inesperadas** entre propuestas que inicialmente parecen
poco viables.

------------------------------------------------------------------------

## **8. Priorización y agrupación de ideas**

Tras la lluvia de ideas, los equipos procedieron a **agrupar y votar las
propuestas**.\
Se presentaron distintos métodos de priorización: - **Votación por
puntos (dot voting)**.\
- **Agrupación por afinidad temática**.\
- **Clasificación en categorías:**\
- *Racionales* (realizables con pocos recursos).\
- *Atractivas pero poco viables*.\
- *Sencillas e inmediatas*.\
- *De largo plazo o innovadoras.*

La combinación de criterios permitió identificar las ideas con **mayor
potencial de desarrollo** para la siguiente fase: **prototipado**.

------------------------------------------------------------------------

## **9. Formulación de la idea de solución**

Cada grupo debía **formular una única idea de solución**, coherente,
innovadora y derivada del proceso anterior.\
Esta idea sería la base para la creación del prototipo en la próxima
sesión.

### **Ejemplos de resultados presentados:**

-   **Educación financiera:**\
    *App tipo juego (similar a Duolingo o Tamagotchi) que incentiva el
    ahorro mediante un asistente virtual personalizable que guía la
    gestión de finanzas de manera lúdica y visual.*\
-   **Cuidado de mascotas:**\
    *Plataforma colaborativa para organizar tareas, recordatorios y
    coordinación entre dueños y cuidadores, garantizando el bienestar
    del animal.*\
-   **Gestión de residuos:**\
    *Aplicación que optimiza rutas de recolección considerando puntos de
    reciclaje y la salud de los recolectores, reduciendo carga física y
    tiempo de trabajo.*

Cada grupo explicó cómo sus ideas respondían a la problemática inicial,
mostrando la coherencia entre empatía, definición y creación de
soluciones.
