## **3. ¿Qué es Design Thinking?**

El **Design Thinking** fue presentado como una **metodología de
innovación y diseño de soluciones** que combina creatividad, análisis y
empatía.\
Su objetivo principal es **comprender a fondo las necesidades del
usuario** antes de definir el problema o pensar en una solución técnica.

### **Características principales**

-   **Enfoque centrado en las personas.**
-   **Aprendizaje iterativo:** se aprende haciendo y probando.\
-   **Equilibrio entre empatía, lógica y creatividad.**
-   **Aplicable a productos, servicios o proyectos sociales.**

### **Ejemplo práctico**

Se revisó el caso de la **Camelbak**, una mochila con sistema de
hidratación creada para deportistas que necesitaban beber agua sin
detenerse.\
Este ejemplo permitió comprender cómo el proceso de diseño parte de una
**necesidad observada en los usuarios** y no de una idea preconcebida.

## **4. Las cinco etapas del Design Thinking**

El profesor explicó las **cinco fases esenciales** del proceso,
destacando que no son lineales, sino **iterativas y flexibles**:

1.  **Empatizar:**\
    Observar, entrevistar y comprender profundamente las experiencias,
    emociones y necesidades de los usuarios.\
    El propósito es **ver el mundo desde su perspectiva**.

2.  **Definir:**\
    Sintetizar la información obtenida para **formular con claridad el
    problema** que se busca resolver.\
    Debe ser una definición inspiradora y concreta.

3.  **Idear:**\
    Generar una amplia gama de ideas posibles sin juzgar su viabilidad
    inicial.\
    Se valoran la **diversidad y cantidad de ideas** más que la
    perfección.

4.  **Prototipar:**\
    Convertir ideas en representaciones tangibles o digitales para poder
    explorarlas y mejorarlas.\
    Los prototipos pueden ser bocetos, maquetas o simulaciones.

5.  **Evaluar:**\
    Probar las soluciones con usuarios reales para obtener
    retroalimentación.\
    Esta etapa alimenta un ciclo de mejora continua.

## **5. Enfoque en las primeras etapas: Empatizar y Definir**

Durante la clase, se profundizó en las dos primeras etapas del proceso,
conocidas como **"espacio del problema"**.

### **Empatizar**

-   Implica establecer una **conexión emocional y cognitiva** con el
    usuario.\
-   Se utilizan herramientas como **entrevistas, mapas de empatía y
    observación directa**.\
-   El propósito es descubrir **insights**, es decir, descubrimientos
    significativos que revelan motivaciones o frustraciones.

### **Definir**

-   Consiste en **organizar y sintetizar los hallazgos** del proceso de
    empatía.\
-   El grupo debe redactar una **definición clara del problema**,
    respondiendo:
    -   ¿Qué necesita realmente el usuario?\
    -   ¿Por qué es importante resolver este problema?\
    -   ¿Qué impacto tendría la solución?

Estas etapas fueron aplicadas en una **actividad grupal**, donde los
estudiantes trabajaron sobre un **escenario ficticio** para ejercitar la
identificación de necesidades y problemáticas.

## **7. Herramientas utilizadas**

Durante el ejercicio se mencionaron o aplicaron las siguientes
herramientas: - **Mapa de empatía:** para representar lo que el usuario
ve, piensa, siente y hace.\
- **User Persona:** representación arquetípica del usuario final.\
- **Journey Map:** recorrido del usuario en relación al problema
identificado.\
- **Matriz de hallazgos y necesidades:** para organizar la información
cualitativa.

Estas herramientas permitieron a los equipos **visualizar el contexto**
del usuario y delimitar de manera más precisa los desafíos a resolver.
