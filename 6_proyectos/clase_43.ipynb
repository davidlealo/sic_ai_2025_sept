## 3. Proyectos y ejemplos destacados

### a. Clasificación de texto y lenguaje natural

* Se repasó el uso de los modelos **TF-IDF**, **LDA** y **LSA** para clasificación temática.
* Ejemplo de aplicación: detección de lenguaje asociado a riesgo suicida en conversaciones o foros, combinando clasificación de texto con variables continuas.
* Se explicó cómo combinar estos modelos con embeddings y buscadores semánticos (usando *Hugging Face*).

### b. Análisis de reseñas y limpieza de datos

* Grupos trabajando con reseñas de productos (ej. Samsung Galaxy S10, S24, etc.).
* Se discutió la eliminación de textos en inglés, duplicados o inconsistentes.
* Reflexión sobre detección de reseñas falsas y usuarios repetidos.
* Se recomendó mantener números de modelo (como “S24”) en los textos para conservar contexto técnico.

### c. Modelos de visión computacional

* Ejemplo práctico: **proyecto de clasificación de gallinas** usando *RoboFlow* y *TensorFlow*.

  * Se explicó la captura de imágenes, entrenamiento del modelo y mejora de métricas mediante aumentación de datos (rotar, recortar, transponer).
  * Caso de uso extendido: monitoreo de gallineros con microcontroladores ESP32 para prevenir accidentes o controlar temperatura.

### d. Generación musical con IA

* Proyecto de conversión de archivos **MP3 a MIDI y a partitura**.

  * Problemas técnicos con incompatibilidades entre versiones de *NumPy* en Google Colab.
  * Soluciones propuestas:

    1. Ejecutar el proceso en local con versión controlada de NumPy.
    2. Implementar API propia para realizar la conversión.
    3. Separar el flujo de trabajo (entrenamiento y conversión) en módulos independientes.
* Se discutió el significado de notas, octavas y representación en pentagrama.
* Recomendación: usar canciones simples para mejorar la precisión del modelo.

## 7. Conceptos técnicos mencionados

* **TF-IDF**: frecuencia de término e inversa de frecuencia de documento.
* **LDA/LSA**: modelos de reducción semántica para clasificación de tópicos.
* **Embeddings**: representaciones vectoriales del texto para búsqueda y similitud.
* **RoboFlow + TensorFlow**: herramientas para etiquetar y entrenar modelos de visión.
* **ESP32**: microcontrolador usado en monitoreo con sensores.
* **Conversión MP3 → MIDI → partitura**: extracción de notas y generación de representación musical.
