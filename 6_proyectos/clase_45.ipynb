## 3. Proyectos revisados

### a. **Predicción de riesgo suicida en estudiantes**

* El grupo analiza textos de reflexiones estudiantiles para detectar señales de alerta.
* Se trabaja con modelos de **clasificación binaria y multiclase**.
* Dificultades: exceso de categorías y ruido en los datos.
* Recomendación del profesor:

  * Simplificar el modelo a tres niveles:

    1. Necesita atención inmediata
    2. Pide ayuda
    3. No presenta riesgo
  * Aplicar un **modelo de embeddings (SentenceTransformer)** de *Hugging Face* para mejorar la detección semántica.
  * Explorar combinación de modelos tipo “votación” (ensemble) para mejorar precisión.

---

### b. **Detección de agresiones en video**

* Proyecto orientado a la **identificación de víctimas y victimarios en situaciones violentas**.
* Se utiliza visión computacional para reconocer movimiento y posturas (levantar la mano, torso, cercanía entre personas).
* Mejoras técnicas:

  * Estabilización de etiquetas temporales (mantener detección durante segundos).
  * Implementación de una función que **genera un clip o captura automática** cuando se detectan agresor y víctima simultáneamente.
  * Posibilidad de exportar resultados a **Streamlit o Gradio** como aplicación visual.
* Discusión sobre optimización con **Grid Search** y control del rendimiento del modelo.
* El profesor propone integrar un “selector de segundos” ajustable para mejorar la experiencia del usuario.

---

### c. **Sistema de recomendación de celulares Samsung**

* El grupo realiza *web scraping* de reseñas de productos Samsung (modelos S y A).
* Objetivo: construir un **sistema de recomendación con análisis de sentimientos y lenguaje natural**.
* Técnicas empleadas:

  * Limpieza de datos para eliminar bots y duplicados.
  * Pruebas con modelos **LDA y LSA** para extracción de temas.
  * Uso de **SentenceTransformer** para clasificación de sentimiento y emoción.
  * Exploración de modelos preentrenados en Hugging Face (`sentiment`, `emotion`, `irony`).
* Recomendación: integrar especificaciones técnicas de los dispositivos como atributos adicionales del modelo.

---

### d. **Clasificación de sonidos ambientales**

* Proyecto enfocado en identificar fuentes de ruido predominantes en grabaciones.
* Problema detectado: el modelo funciona bien con sonidos simples, pero falla al procesar secuencias largas o superpuestas.
* Recomendaciones:

  * Segmentar los audios antes del entrenamiento.
  * Probar diferentes representaciones (MFCC, espectrogramas mel).
  * Ajustar los parámetros del modelo antes de retomar el entrenamiento.

---

### e. **Bot en Telegram para detección de objetos en imágenes**

* Proyecto individual de Antonio M.
* El bot recibe fotos enviadas por el usuario y responde con una **clasificación visual automática**.
* Se elige esta vía para simplificar la entrega y conectar la interfaz de IA con un uso práctico inmediato.
* Próximo paso: integrar el modelo entrenado y generar respuestas automatizadas desde Telegram.
