## Presentación del Grupo: Redes Neuronales Convolucionales (CNN)

El grupo explicó que las **CNN** son modelos diseñados principalmente para el **reconocimiento y clasificación de imágenes**, inspiradas en la forma en que **la corteza visual humana** detecta patrones en diferentes niveles (bordes, formas, objetos).

### Aspectos conceptuales abordados:

1. **Convolución**

   * Se aplica un **kernel o filtro** sobre una imagen para resaltar ciertos patrones.
   * Se mostró la diferencia entre convolución en continua (integral) y discreta (suma).
   * Se explicó cómo el filtro se desplaza sobre la imagen para producir un **mapa de características**.

2. **Stride y Padding**

   * **Stride:** número de posiciones que el filtro se mueve en cada paso.
   * **Padding:** agregado de bordes con ceros para evitar que la imagen se reduzca.

3. **Pooling (Agrupación)**

   * Se revisó el **Max Pooling** para reducir dimensiones sin perder rasgos clave.
   * Se explicó cómo esta técnica hace más eficiente el modelo.

4. **Estructura general de una CNN**

   * Imagen de entrada
   * Capas de extracción de características (convolución + pooling)
   * Aplanamiento (**Flatten**)
   * Capas densas de clasificación
   * Capa de salida con predicción (ej: clasificación de dígitos del 0 al 9)
